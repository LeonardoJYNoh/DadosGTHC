# Exercício 2
Mostre as primeiras 10 linhas somente com as colunas Nome do Anúncio, Nome do Anfitrião, Pais e Preço, filtrando apenas quartos do tipo 'Private room'

```python
query = '''
SELECT
  NAME,
  HOST_NAME,
  COUNTRY,
  PRICE
FROM GTHC_AIRBNB_SOR
WHERE
  ROOM_TYPE = 'Private room'
LIMIT 10;
'''

pd.read_sql(query, conn)
```

# Exercício 3
Qual os TOP 5 tipos de quarto com maior quantidade de anúncios?
Sua query deve retornar apenas os primeiros 5 tipos de quarto com a respectiva quantidade.

```python
query = '''
SELECT
  ROOM_TYPE,
  COUNT(*) AS QUANTIDADE
FROM GTHC_AIRBNB_SOR
GROUP BY ROOM_TYPE
ORDER BY QUANTIDADE DESC
LIMIT 5;
'''

pd.read_sql(query, conn)
```

# Exercício 4
Qual a quantidade e ano de construção médio dos anúncios agrupados por pais, filtrando apenas aqueles que tem quartos privados e a nota média de avaliação maior que 3?


```python
query = '''
SELECT
  COUNTRY,
  COUNT(*) AS QUANTIDADE,
  AVG(CAST(CONSTRUCTION_YEAR AS INT)) AS MEDIA_ANO_CONSTRUCAO
FROM GTHC_AIRBNB_SOR
WHERE
  REVIEW_RATE_NUMBER > 3
  AND ROOM_TYPE = 'Private room'
GROUP BY COUNTRY
'''

pd.read_sql(query, conn)
```

# Exercício 5 - Desafio!
Qual a quantidade total, quantidade distinta e soma do valor dos anúncions localizados nos Estados Unidos, onde a quantidade total de avaliações é maior que 5 e o nome do anfitrição tem a letra 'a'?  
  
_utilize o que aprendeu em aula, mas também pesquise na internet_

```python
query = '''
SELECT
  COUNTRY,
  COUNT(*) AS QUANTIDADE_TOTAL,
  COUNT(DISTINCT PRICE) AS QUANTIDADE_DISTINCT,
  SUM(CAST(REPLACE(REPLACE(PRICE, '$', ''), ',', '')AS INT)) AS PRECO

FROM GTHC_AIRBNB_SOR
WHERE
  COUNTRY = 'United States'
  AND PRICE != 'None'
  AND CAST(NUMBER_OF_REVIEWS AS INT) > 5
  AND HOST_NAME ~* 'A'
GROUP BY COUNTRY

'''

pd.read_sql(query, conn)
```
